# Dictionaries

Define - Dictionaries are ordered collection of data items. They store multiple items in a single variable. Dictionary items are key-value pairs that are separated by commas and enclosed within curly brackets {}.

Key Characterstics -

A collection of key value Pair eg. {Key:Value} pair

Ordered and Changeble and NO Duplicates

```python
# Example

info = {'name':'Karan', 'age':19, 'eligible':True}
print(info)
```

## Difference between Set and Dictionary

| Feature | Set | Dictionary |
|---------|-----|------------|
| **Structure** | Unordered unique elements | Key-value pairs |
| **Duplicates** | No duplicates allowed[attached_file:1] | Keys unique, values can duplicate[attached_file:1] |
| **Access** | Membership testing (`in`)[web:3] | Key lookup (`dict[key]`)[web:3] |
| **Syntax** | `{1, 2, 3}` or `set()`[attached_file:1] | `{'a': 1, 'b': 2}` or `dict()`[attached_file:1] |
| **Mutable** | Yes (add/remove elements)[attached_file:1] | Yes (add/remove/modify pairs)[attached_file:1] |
| **Ordered** | Unordered (pre-3.7), insertion order (3.7+)[attached_file:1] | Insertion order since Python 3.7[attached_file:1] |
