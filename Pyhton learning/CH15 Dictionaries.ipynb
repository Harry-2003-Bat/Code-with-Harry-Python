# Dictionaries

Define - Dictionaries are ordered collection of data items. They store multiple items in a single variable. Dictionary items are key-value pairs that are separated by commas and enclosed within curly brackets { }.

Key Characterstics -

A collection of key value Pair eg. {Key:Value} pair

Ordered and Changeble and NO Duplicates

```python
# Example

info = {'name':'Karan', 'age':19, 'eligible':True}
print(info)
```

```python
Dic = {
    "Name": "Harry",
    "Age": 21,
    "Gender ":"Male"
}
print(Dic["Name"])                # Access the Dictionary item
```

## Difference between Set and Dictionary



| Feature | Set | Dictionary |
|---------|-----|------------|
| **Structure** | Unordered unique elements | Key-value pairs |
| **Duplicates** | No duplicates allowed | Keys unique, values can duplicate |
| **Access** | Membership testing (`in`) | Key lookup (`dict[key]`) |
| **Syntax** | `{1, 2, 3}` or `set()` | `{'a': 1, 'b': 2}` or `dict()` |
| **Mutable** | Yes (add/remove elements) | Yes (add/remove/modify pairs) |
| **Ordered** | Unordered (pre-3.7), insertion order (3.7+) | Insertion order since Python 3.7 |






## Accessing Dictionary items 

### Accessing Single Value :

Values in a dictionary can be accessed using keys. We can access dictionary values by mentioning keys either in square brackets or by using get method.

```python
info = {'name':'Karan', 'age':19, 'eligible':True}
print(info['name'])

print(info.get("age"))                 # get() is safe access Agar key exist karti hai → value return karega
print(info.get('eligible'))            # info.get() - Agar key exist nahi karti → ❌ error nahi dega, None return karega 
```

### Accessing keys :

We can print all the keys in the dictionary using keys() method.

```python
info = {'name':'Karan', 'age':19, 'eligible':True}
print(info.keys())
```

### Accessing multiple values :

We can print all the values in the dictionary using values() method.

```python
info = {'name':'Karan', 'age':19, 'eligible':True}
print(info.values())
```

### Accessing key-value pairs :

We can print all the key-value pairs in the dictionary using items() method.

```python
info = {'name':'Karan', 'age':19, 'eligible':True}
print(info.items())
```

### update() -

The update() method updates the value of the key provided to it if the item already exists in the dictionary, else it creates a new key-value pair.

```python
info = {'name':'Karan', 'age':19, 'eligible':True}
print(info)
info.update({'age':20})
info.update({'DOB':2001})
print(info)
```

### Remove() -

the remove() method remove the value in the existing dictionary 

```python
a = {1,2,3,4,5,8}
a.remove(8)
print(a)
```

### clear() -
The clear() method removes all the items from the list.


```python
a = {1,2,3,4,4,5,5,6,6,6}
a.clear()
print(a)
```

### pop() -

The pop() method removes the key-value pair whose key is passed as a parameter.

```python
a = {"Name":"Harry", "Age": 21, "Eligible" : "Yes" }
a.pop("Eligible")
a.pop("Age")
print(a)
```

### popitem() -

The popitem() method removes the last key-value pair from the dictionary.

```python
info = {'name':'Karan', 'age':19, 'eligible':True, 'DOB':2003}
info.popitem()
print(info)
```
