
# Function In Python

```python
# A function is a block of code that performs a specific task whenever it is called

# Function - There are two types of functions: codeword for large block of code.

# Built-in functions - These functions are defined and pre-coded in python. Some examples of built-in functions are as follows:
#                      min(), max(), len(), sum(), type(), range(), dict(), list(), tuple(), set(), print(), etc.

# User-defined functions - We can create functions to perform specific tasks as per our needs. 
#                          Such functions are called user-defined functions.
```

```python
# Syntax - def function_name(parameters): 
# Code and Statements

# Create a function using the def keyword, followed by a function name, followed by a paranthesis (()) and a colon(:).
# Any parameters and arguments should be placed within the parentheses.
# Rules to naming function are similar to that of naming variables.
# Any statements and other code within the function should be indented.
```

***Function Argument In Python***

```python
#TODO Function Argumnet - There are four types of arguments that we can provide in a function:

# Default Arguments
# Keyword Arguments
# Variable length Arguments
# Required Arguments
```

```python
# Default argumnet - Provide a default value in the function definition. 
#                  - If no value is passed during the function call, the default is used.

# Example - yha humne last mein name("Harsh") ek default argument hai joh humne last mein pass kari hai

def name(fname, mname = "wardhan", lname ="Jadho"):
    print("Hello,",fname, mname, lname)
name("Harsh")  
```

```python
# Keyword Arguments: Pass arguments using key=value syntax. Random oreder mein output le sakte hai random order
# Example - 

def name(fname, mname, lname):
    print("Hello", fname,mname,lname)
name(mname="Tony",fname="Harvard",lname="Stark")
```

```python
# Variable Length Argument - yha *name ko tuple de diya jisse aap jitne chao utne naam le sakte ho

def name(*name):
    print("Hello,", name[0], name[1], name[2])

name("James", "Buchanan", "Barnes")
```

```python
# Return Statement - The return statement is used to return the value of the expression back to the calling function.

def name(fname, mname, lname):
    return "Hello, " + fname + " " + mname + " " + lname

print(name("James", "Buchanan", "Barnes"))
```

## Enumerate function:

Bas bhai, simple shabdon mein kahein toh enumerate() Python ka ek aisa shortcut hai jo aapko list ke items ke saath-saath unka index (position) bhi de deta hai.

```python
for rank, name in enumerate(['Rahul', 'Sonia', 'Amit'], start=1):
    print(f"Rank {rank}: {name}")
```

```python
# Example of enumerate with a list
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits, start=0):
	print(index, fruit)
```


# Python Variable Scope: Local vs. Global

| Feature | Local Variable | Global Variable |
| :--- | :--- | :--- |
| **Where it’s defined** | Inside a function | Outside all functions |
| **Lifespan** | Dies when function ends | Lives until the program stops |
| **Visibility** | Only inside that function | Everywhere in the file |

### Local Variable 

A local variable is defined inside a function. It only exists while that function is running. Once the function finishes, the variable is "thrown away."

Access: Only accessible inside that specific function.

Analogy: Like a book on your nightstand. You can use it while you're in your bedroom, but someone in the kitchen can’t see it or use it.



```python
def room():
    item=("table","book","chair")   # This ia a Local variable
    print(item)
room()
```

### Global Variable 

A global variable is defined outside of any function, usually at the very top of your script.

Access: Accessible from anywhere in your code—inside or outside of functions.

Analogy: Like a giant statue on your front lawn. Anyone looking through any window of the house (any function) can see it.

```python
msg = "Hello scenario"  # This is a global variable

def greeting():
    print(msg)  # The function can see the global variable

greeting()
print(msg)      # This works too!
```
