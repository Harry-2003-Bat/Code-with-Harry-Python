Modules aur Importing ka Introduction - Python mein import ka matlab hai kisi dusri Python file (module) ke code ko apni file mein load karna. Isse aapko wahi code baar-baar likhne ki zaroorat nahi padti.

Key Points:

Jab aap koi module import karte ho, toh Python us file ka saara code execute karta hai.

Iske baad us module ke saare functions aur variables aapke program mein use ke liye ready ho jaate hain.

````python
import math

result = math.sqrt(16)
print(f"16 ka square root hai: {result}")
````

## from Keyword ka Use - 
Agar aapko poore module ki zaroorat nahi hai aur sirf kuch specific functions chahiye, toh from ka use karein.

````python
# Importing specific attributes
from math import sqrt, pi

# No need to use 'math.' prefix
print (sqrt(9))
print (pi)
````

## Using the as Keyword (Aliasing) -

The as keyword is used to create an alias (a shorter name) for a module or a function. This is very common in Data Science (e.g., pandas as pd).

````python
# Aliasing a module
import math as m
import pandas as pd

print(m.sqrt(25))

# Aliasing a specific function
from math import factorial as fact
print(fact(5)) 
````

## The dir() Function -
The dir() function is a built-in Python tool that helps you see all the functions, variables, and classes defined inside a module.

````python
import math

# List all attributes of the math module
print(dir(math))
````

# `if __name__ == "__main__"` Python mein Samajhna

`if __name__ == "__main__"` ek common Python idiom hai jo yeh check karne ke liye use hota hai ki ek Python script directly run ho rahi hai ya phir kisi module ke roop mein import ki gayi hai. Yahaan iska purpose aur usage samjha gaya hai:

### Iska Matlab Kya Hai?
- Har Python module mein ek special built-in attribute hota hai jise `__name__` kehte hain.
- Jab ek Python file directly run hoti hai, toh `__name__` attribute ki value `"__main__"` hoti hai.
- Jab ek Python file ko ek module ke roop mein import kiya jata hai, toh `__name__` attribute ki value us module ke naam ke barabar hoti hai.

### Iska Use Kyun Karein?
Yeh construct aapko yeh karne deta hai:
1. Aapka code reuse hone layak banata hai jab aap usse kisi aur script mein module ke roop mein use karte ho.
2. Code execution ko alag karta hai (jab script directly run hoti hai) aur functions, classes ya doosre reusable components ke definition se.

### Example
```python
# my_module.py

def greet():
    print("Module se Hello!")

if __name__ == "__main__":
    print("Yeh script directly run ho rahi hai.")
    greet()
else:
    print("Yeh script ek module ke roop mein import ki gayi hai.")
```

### Example Ka Explanation
1. Jab aap `python my_module.py` run karte ho:
   - `if` condition `True` hoti hai.
   - Script "Yeh script directly run ho rahi hai." print karti hai aur `greet()` function ko call karti hai.

2. Jab aap `my_module` ko kisi doosri script mein import karte ho:
   - `if` condition `False` hoti hai.
   - Script `if` block ke andar ka code execute nahi karti.

### Best Practices
- `if __name__ == "__main__"` ka use test cases ya script-specific logic ke liye karein.
- Reusable code (functions, classes) ko `if` block ke bahar rakhein taaki module import-friendly rahe.
