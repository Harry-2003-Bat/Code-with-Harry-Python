# SET

Define - Sets are collections of well-defined, non-repeated objects in Python, They are designed to prevent repeated entries, ensuring each item is unique and used Semi column for sepration and closed with curly bracket

Key characteristics -

Unordered (no indexing or slicing)

No duplicate elements

Mutable (you can add/remove items)

Elements must be hashable (e.g., numbers, strings, tuples)

```python
s = {1,"Harry",84.8,"548R5"}
print(s)               # Unordered (no indexing or slicing)
```

```python
s =  {1,2,3,4,5,5,5,5}
print(s)                # No Duplicate element 
```

```python
# Creating a set with hashable elements

my_set = {1, "apple", (10, 20)}    # Elements must be hashable It has a fixed (unchangeable) value during its lifetime
                                   # hash(10)        # valid          
                                   # hash("apple")   # valid          
                                   # #hash((1, 2))    # valid

print(my_set)         
              
# Add an element

my_set.add(5)

# Remove an element

my_set.remove("apple")

print(my_set)           # Mutable (you can add/remove items)
```

## Set Mutation 

### Union() : 


The union (Dono ko Merge kar deta hai) The union() method returns a new set 

```python
A = {1,2,3}                                    
B = {4,5,6}
C = A.union(B)
print(C)
```

### Update() :

update (Ek set ki value dusre set mein daal deta hai) method adds item into the existing set from another set.

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
A.update(B)
print(A)
```

### Intersection() :

The intersection() methods prints only items that are similar to both the sets. 

```python
A = {1,2,3}
B = {2,3,4}
C = A.intersection(B)
print(C)
```

### Symmetric_difference() :

The symmetric_difference() prints only items that are not similar to both the sets

```python
A = {1,2,3,4}
B = {0,1,2,3}
C = A.symmetric_difference(B)
print(C)
```

### Difference() : 

The difference() prints only items that are only present in the original set and not in both the sets.

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = A.difference(B)
print(C)
```

### Isdisjoint() :

Disjoint sets are sets that have no common elements This method returns False if items are present, else it returns True.

```python
A = {1,2,3}                        # sets are not disjoint because 3 is common.
B = {3,4,5}
print(A.isdisjoint(B))

C = {1,2,3,4,5}                    # C and D are disjoint sets because they share no elements.
D = {6,7,8,9,10}
print(C.isdisjoint(D))
```

### Add() :

If you want to add a single item to the set use the add() method.

```python
A = {1,2,3}
A.add(4)
print(A)
```

### Remove() :
We can use remove() and discard() methods to remove items form list.

```python
A = {1,2,3,4}
A.remove(4)
print(A)
```

### Pop : 

This method removes the last item of the set but the catch is that we donâ€™t know which item gets popped as sets are unordered. However, you can access the popped item if you assign the pop() method to a variable.


```python
A = {1,2,3,4,5}
B = A.pop()

print(A)
print(B)
```

### Del :

del is not a method, rather it is a keyword which deletes the set entirely.

```python
A = {1,2,3,4}
del A
print(A)
```

### Check if item exists :

You can also check if an item exists in the set or not.

```python
A = {1,2,3,4,5}
if 1 in A:
    print("1 is Present")
else:
    print("1 is Absent")
```
