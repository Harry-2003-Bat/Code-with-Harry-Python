***Recursion in Python*** 

~ in recursion the function call itself to solve the bigger problem.
 
~ Recursion means a function solving a big problem by repeatedly solving smaller versions of the same problem until it reaches a very easy case and stops.

Example - Factorial series

```python
# Find the factorial of 5

def factorial (n):
    if ( n == 0 or n == 1):
        return 1
    else:
        return n * factorial (n-1)

print(factorial(5))

# Logic behind it - 

# 5 * factorial(4)
# 5 * 4 * factorial(3)
# 5 * 4 * 3 * factorial(2)
# 5 * 4 * 3 * 2 * factorial(1)
# 5 * 4 * 3 * 2 * 1
```

```python
# Fibonacci Series using Recursion
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Example usage
n = 10  # Number of terms in the Fibonacci series
for i in range(n):
    print(fibonacci(i), end=" ")
```
