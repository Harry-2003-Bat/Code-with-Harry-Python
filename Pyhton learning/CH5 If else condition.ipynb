### Notes on Python `time` Module

The `time` module in Python provides various time-related functions to work with time in your programs. Below is a detailed explanation of the `time` module:

#### Commonly Used Functions in `time` Module:

1. **`time.time()`**
   - Returns the current time in seconds since the epoch (January 1, 1970, 00:00:00 UTC).
   - Example:
     ```python
     import time
     current_time = time.time()
     print("Current Time (in seconds):", current_time)
     ```

2. **`time.sleep(seconds)`**
   - Suspends (pauses) the execution of the current thread for the specified number of seconds.
   - Example:
     ```python
     import time
     print("Start")
     time.sleep(2)  # Pauses for 2 seconds
     print("End")
     ```

3. **`time.localtime([secs])`**
   - Converts a time expressed in seconds since the epoch to a `struct_time` in local time.
   - If no argument is provided, the current time is used.
   - Example:
     ```python
     import time
     local_time = time.localtime()
     print("Local Time:", local_time)
     ```

4. **`time.gmtime([secs])`**
   - Converts a time expressed in seconds since the epoch to a `struct_time` in UTC.
   - Example:
     ```python
     import time
     utc_time = time.gmtime()
     print("UTC Time:", utc_time)
     ```

5. **`time.strftime(format[, t])`**
   - Converts a `struct_time` to a string as per the given format.
   - Example:
     ```python
     import time
     formatted_time = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())
     print("Formatted Time:", formatted_time)
     ```

6. **`time.strptime(string, format)`**
   - Parses a string representing time and returns a `struct_time` object.
   - Example:
     ```python
     import time
     time_obj = time.strptime("2026-02-16", "%Y-%m-%d")
     print("Parsed Time:", time_obj)
     ```

7. **`time.ctime([secs])`**
   - Converts a time expressed in seconds since the epoch to a string representing local time.
   - Example:
     ```python
     import time
     readable_time = time.ctime()
     print("Readable Time:", readable_time)
     ```

8. **`time.perf_counter()`**
   - Returns the value of a high-resolution performance counter (useful for measuring time intervals).
   - Example:
     ```python
     import time
     start = time.perf_counter()
     # Some code execution
     end = time.perf_counter()
     print("Execution Time:", end - start)
     ```

9. **`time.process_time()`**
   - Returns the CPU time used by the current process.
   - Example:
     ```python
     import time
     cpu_time = time.process_time()
     print("CPU Time:", cpu_time)
     ```

#### `struct_time` Object:
The `struct_time` object is a tuple-like object with the following attributes:
- `tm_year`: Year (e.g., 2026)
- `tm_mon`: Month (1-12)
- `tm_mday`: Day of the month (1-31)
- `tm_hour`: Hour (0-23)
- `tm_min`: Minute (0-59)
- `tm_sec`: Second (0-61, including leap seconds)
- `tm_wday`: Day of the week (0 is Monday, 6 is Sunday)
- `tm_yday`: Day of the year (1-366)
- `tm_isdst`: Daylight Saving Time flag (-1, 0, 1)

#### Example of `struct_time`:
```python
import time
local_time = time.localtime()
print("Year:", local_time.tm_year)
print("Month:", local_time.tm_mon)
print("Day:", local_time.tm_mday)
```

#### Notes:
- The `time` module is useful for working with timestamps, delays, and performance measurements.
- For more advanced date and time manipulations, consider using the `datetime` module.
