***Recursion in Python*** 

~ in recursion the function call itself to solve the bigger problem.
 
~ Recursion means a function solving a big problem by repeatedly solving smaller versions of the same problem until it reaches a very easy case and stops.

Example - Factorial series

```python
# Find the factorial of 3

def factorial (n):
    if ( n == 0 or n == 1):
        return 1
    else:
        return n * factorial (n-1)

print(factorial(5))

# Logic behind it - 

# 5 * factorial(4)
# 5 * 4 * factorial(3)
# 5 * 4 * 3 * factorial(2)
# 5 * 4 * 3 * 2 * factorial(1)
# 5 * 4 * 3 * 2 * 1
```

```python
# 0 1 1 2 3 5 8 
def fibo (n):
    if ( n == 0):
        return 0
    if ( n == 1 ):
        return 1
    else:
        return fibo(n) + n-1
    
print(fibo(n))
```

```python
def fibo(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fibo(n - 1) + fibo(n - 2)

# Example usage
n = 5  # Replace with the desired value of n
print(fibo(n))
```
